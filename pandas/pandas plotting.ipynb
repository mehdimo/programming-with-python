# Pandas Plots

```python
from numpy import random

x = random.choice(range(18,90), 100)

data = data = {
 'Id': range(100, 200),
 'Age': random.choice(range(18,90), 100), 
 'State': random.choice(['TX','DC','CA','MI'], 100),
 'Balance':  random.choice(range(10000), 100)
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df["Balance"].plot()
```

```python
df.plot(kind="scatter", x="Age", y="Balance", color="red")
```

```python
df["Balance"].plot(kind="hist")
```

```python
df["Age"].plot(kind="hist", color="green")
```
